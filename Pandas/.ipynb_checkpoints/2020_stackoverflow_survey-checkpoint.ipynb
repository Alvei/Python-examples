```python
# inspired by https://github.com/CoreyMSchafer/code_snippets/blob/master/Python/Pandas/ 
# data from https://insights.stackoverflow.com/survey
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
sns.set_style('whitegrid')

pd.set_option('display.max_columns', 85)
pd.set_option('display.max_rows', 85)
#pd.options.display.float_format = "{:,.2f}".format
```

```python
data_type = {'Hobbyist': "category", "CompFreq": "category", "JobSat": "category", "SOAccount": "category", "Trans": "category"}
```

```python
df = pd.read_csv('data/survey_results_public.csv', index_col='Respondent', dtype = data_type)
schema_df = pd.read_csv('data/survey_results_schema.csv', index_col='Column')
df.shape
```

```python
# Notice that the age contains numbers and string. 
# df['Age1stCode'].apply(type).value_counts()
df['Age1stCode'] = pd.to_numeric(df['Age1stCode'], errors='coerce')
df['WorkWeekHrs'] = pd.to_numeric(df['WorkWeekHrs'], errors='coerce')
df['YearsCode'] = pd.to_numeric(df['YearsCode'], errors='coerce')
df['YearsCodePro'] = pd.to_numeric(df['YearsCodePro'], errors='coerce')
```

```python
df.rename(columns={'ConvertedComp': 'SalaryUSD'}, inplace=True)
```

## Convert some categorical

```python
# Convert Yes/No to True/False leaving NaN untouched
df['Hobbyist'] = df['Hobbyist'].map({'Yes': True, 'No': False})
df['Trans'] = df['Trans'].map({'Yes': True, 'No': False})
df['SOAccount'] = df['SOAccount'].map({'Yes': True, 'No': False})
```

```python
YearsCode = df['YearsCode'].value_counts().sort_index()
ax = sns.barplot(YearsCode.index, YearsCode)
ax.set(xlabel='Years coding', ylabel='Number of respondent')
x = YearsCode.index
#plt.xticks(np.arange(min(x), max(x)+1, 10.0))
start, end = ax.get_xlim()
stepsize=10
ax.xaxis.set_ticks(np.arange(start, end, stepsize))
ax.set_xticklabels([0, 10, 20, 30, 40])
plt.show()
```
